# LINEAR REGRESSION MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
model=LinearRegression()
tp=India.select_dtypes(np.number)
X=tp.drop(['total_cases'],axis=1)
y=tp['total_cases']
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.50,random_state=42)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print('Accuracy of Linear Regression Model is : {}'.format(model.score(X_test,y_test)))
```
